```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define a structure cpntains PMT number, Darkrate, Darkrate_err, Testsheet_darkrate, Meanarea_uw, Meanarea_uw_err, Meanarea_db, Meanarea_db_err

class PMT:
    def __init__(self, number, Darkrate, Darkrate_err, Testsheet_darkrate, Meanarea_uw, Meanarea_uw_err, Meanarea_db, sigma_uw):
        self.number = number
        self.Darkrate = Darkrate
        self.Darkrate_err = Darkrate_err
        self.Testsheet_darkrate = Testsheet_darkrate
        self.Meanarea_uw = Meanarea_uw
        self.Meanarea_uw_err = Meanarea_uw_err
        self.Meanarea_db = Meanarea_db
        # self.Meanarea_db_err = Meanarea_db_err
        self.sigma_uw = sigma_uw
```

```python
PMT_num_list = [1123, 1115, 1107, 1119, 1111, 1114, 1110, 1118, 1125, 1109, 1108, 1112, 1113,1117]
Darkrate_list = [806, 1017, 838, 1483, 1068, 1165, 1448, 811, 1745, 1233, 1188, 906, 898, 855]
Darkrate_err_list = [4, 4, 3, 2, 1, 1, 1, 1, 2, 1, 1, 1, 1, 1]
Testsheet_darkrate_list = [1200, 900, 1200, 1200, 1000, 1400, 1100, 1900, 1500, 1500, 1800, 1300, 800, 800]

Meanarea_uw_list = [114, 168, 138, 168, 163, 166, 159, 147, 179, 155, 151, 121, 135, 88]
Meanarea_uw_err_list = [0.39, 0.69, 0.29, 0.21, 0.23, 0.22, 0.24, 0.23, 0.142, 0.26, 0.30, 0.12, 0.3, 0.23]
Meanarea_db_list = [117, 159, 159, 159, 159, 159, 159, 159, 159, 159, 159, 159, 159, 63]
# Meanarea_db_err_list = [20, 66, 29, 36, 30, 40, 52, 34, 49, 41]

sigma_uw_list = [20, 51, 28, 45, 47, 49, 46, 37, 46, 42, 40, 20, 28, 13]
optimal_voltage = [1978, 1900, 1822, 1819, 1792, 1774, 1756, 1732, 1715, 1706, 1704, 1643, 1608, 2204]
```

```python
PMT_uw_data = []
for i in range(len(PMT_num_list)):
    PMT_uw_data.append(PMT(PMT_num_list[i], Darkrate_list[i], Darkrate_err_list[i], Testsheet_darkrate_list[i], Meanarea_uw_list[i], Meanarea_uw_err_list[i], Meanarea_db_list[i], sigma_uw_list[i]))
```

```python
#sort PMT_uw_data by PMT number
PMT_uw_data.sort(key=lambda x: x.number)
```

```python
#plot Darkrate and Testsheet_darkrate
PMT_num = [PMT_uw_data[i].number for i in range(len(PMT_uw_data))]
Darkrate = [PMT_uw_data[i].Darkrate for i in range(len(PMT_uw_data))]
Darkrate_err = [PMT_uw_data[i].Darkrate_err for i in range(len(PMT_uw_data))]
Testsheet_darkrate = [PMT_uw_data[i].Testsheet_darkrate for i in range(len(PMT_uw_data))]

x_pos = np.arange(len(PMT_num)) + 1
print(x_pos)

plt.figure(figsize=(10, 5))
plt.errorbar(x_pos, Darkrate, yerr=Darkrate_err, fmt='o', label='Unserwater test')
plt.plot(x_pos, Testsheet_darkrate, 'o', label='Testsheet')
plt.xlabel('PMT number')
plt.xticks(x_pos, PMT_num)
plt.ylabel('Dark Rate (Hz)')
plt.ylim(0, 2000)
plt.legend()
plt.title('Underwater Test and Testsheet Dark Rate')
```

```python
#plot Meanarea_uw and Meanarea_db
Meanarea_uw = [PMT_uw_data[i].Meanarea_uw for i in range(len(PMT_uw_data))]
Meanarea_uw_err = [PMT_uw_data[i].Meanarea_uw_err for i in range(len(PMT_uw_data))]
Meanarea_db = [PMT_uw_data[i].Meanarea_db for i in range(len(PMT_uw_data))]
# Meanarea_db_err = [PMT_uw_data[i].Meanarea_db_err for i in range(len(PMT_uw_data))]

x_pos = np.arange(len(PMT_num)) + 1

plt.figure(figsize=(10, 5))
plt.errorbar(x_pos, Meanarea_uw, yerr=Meanarea_uw_err, fmt='o', label='Underwater test', color='red', capsize=5, capthick=2, ecolor='red')
# plt.errorbar(x_pos, Meanarea_db, yerr=Meanarea_db_err, fmt='o', label='Darkbox test', color='blue', capsize=5, capthick=2, ecolor='blue')
plt.plot(x_pos, Meanarea_db, 'o', label='Darkbox test', color='blue')
plt.xlabel('PMT number')
plt.xticks(x_pos, PMT_num)
plt.ylabel('Mean Area')
plt.ylim(0, 200)
plt.legend()
plt.title('Underwater Test and Darkbox Test Mean Area')
```

```python
sigma_mu = [PMT_uw_data[i].sigma_uw/PMT_uw_data[i].Meanarea_uw for i in range(len(PMT_uw_data))]
print(sigma_mu)
plt.figure(figsize=(10, 5))
plt.plot(x_pos, sigma_mu, 'o', label='Underwater test', color='red')
plt.xlabel('PMT number')
plt.xticks(x_pos, PMT_num)
plt.ylabel('$\sigma/\mu$')
plt.ylim(0, 0.5)
plt.legend()
plt.title('Resolution in Underwater Test')
```
